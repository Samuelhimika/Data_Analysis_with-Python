```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
x= r"C:\Users\TRUTH & JUSTICE\Desktop\Dataset1.csv"
data= pd.read_csv(x)
data.head()
```

```python
sales_data = data['Price'] 

# Create a histogram
plt.figure(figsize=(10, 10))
plt.hist(sales_data, bins=30, color='orange', edgecolor='white')
plt.title('Histogram showing Quantity against price')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()
```

```python
sales_data = data[['Product', 'Price']]

# Create a box plot
plt.figure(figsize=(6, 6))
sns.boxplot(x='Product', y='Price', data = sales_data)
plt.title('Box Plot of Price by Product')
plt.xlabel('Product')
plt.ylabel('Price')
plt.show()
```
